```python
import numpy as np
from sklearn.metrics import confusion_matrix, precision_score, recall_score, accuracy_score
from sklearn.metrics import classification_report, roc_curve, auc
import matplotlib.pyplot as plt

# Actual and Predicted labels
actual = ['P','P','N','N','P','N','N','N','P','N','N','N','N','P','N','P','P','P','P','N','P','N','N','P']
predicted = ['P','N','N','N','P','N','N','N','P','P','N','N','N','P','N','P','P','P','P','P','P','P','N','P']

# Convert labels to numeric (P=1, N=0)
y_true = [1 if x=='P' else 0 for x in actual]
y_pred = [1 if x=='P' else 0 for x in predicted]

# 1. Confusion Matrix
cm = confusion_matrix(y_true, y_pred)
print("Confusion Matrix:")
print(cm)

# 2. Precision
precision = precision_score(y_true, y_pred)
print("\nPrecision:", precision)

# 3. Recall
recall = recall_score(y_true, y_pred)
print("Recall:", recall)

# 4. Accuracy
accuracy = accuracy_score(y_true, y_pred)
print("Accuracy:", accuracy)

# 5. Classification Report
print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=['Negative','Positive']))

# 6. ROC Curve and AUC
fpr, tpr, thresholds = roc_curve(y_true, y_pred)
roc_auc = auc(fpr, tpr)

print("AUC:", roc_auc)

# 7. Draw ROC Curve
plt.figure()
plt.plot(fpr, tpr, label="ROC curve (AUC = %0.2f)" % roc_auc)
plt.plot([0,1], [0,1], linestyle='--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend(loc="lower right")
plt.show()
```

7. Apply Decision tree algorithm using entropy (Information gain) to develop classifier model using 80/20 
data split on IRIS dataset 
* Develop classifier model by splitting the dataset to 70-30 of training set and testing set respectively 
* Generate the developed tree 
* Represent the developed tree in terms of text 
* Display the classification report that shows the performance of the classifier model tested by the 
testing dataset 
* Generate the confusion matrix without class label 
* Generate the confusion matrix by incorporating class label 
* Calculate the accuracy, precision, recall and F-measure  
* Find the class label of the record with the value of {6.4, 1.8, 6.6, 2.1} 

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score

# Load dataset with column names
file_path = r"C:\Users\yifru\OneDrive\Desktop\machine learning\worksheet four\iris\iris.data"

data = pd.read_csv(file_path, header=None,
                   names=['SepLength','SepWidth','PetLength','PetWidth','Class'])

# Features and target
X = data.iloc[:,0:4]
Y = data['Class']

# Split dataset (70/30)
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=100)

# Decision Tree with entropy
clf = DecisionTreeClassifier(criterion="entropy", random_state=42)
clf.fit(X_train, y_train)

# Predict class of new record
sample = [[6.4,1.8,6.6,2.1]]
prediction = clf.predict(sample)
print("Predicted class for {6.4,1.8,6.6,2.1}:", prediction)

# Predict test dataset
y_pred = clf.predict(X_test)

# Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:\n", conf_matrix)

# Accuracy
print("\nAccuracy:", accuracy_score(y_test, y_pred))

# Precision, Recall, F1
precision = precision_score(y_test, y_pred, average='macro')
recall = recall_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')

print("\nPrecision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

8. Apply Decision tree using gini index to develop classifier model using 80/20 data split on IRIS dataset 
* Develop classifier model by splitting the dataset to 70-30 of training set and testing set respectively 
* Generate the developed tree 
* Represent the developed tree in terms of text 
* Display the classification report that shows the performance of the classifier model tested by the 
testing dataset 
* Generate the confusion matrix without class label 
* Generate the confusion matrix by incorporating class label 
* Calculate the accuracy, precision, recall and F-measure  
* Find the class label of the record with the value of {6.4, 1.8, 6.6, 2.1} 

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, export_text
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# Load Iris dataset with column names
file_path = r"C:\Users\yifru\OneDrive\Desktop\machine learning\worksheet four\iris\iris.data"

data = pd.read_csv(file_path, header=None,
                   names=['SepLength','SepWidth','PetLength','PetWidth','Class'])

# Features and target
X = data.iloc[:,0:4]
Y = data['Class']

# Split dataset (70% training, 30% testing)
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=100)

# Decision Tree using Gini index
clf = DecisionTreeClassifier(criterion="gini", random_state=42)
clf.fit(X_train, y_train)

# Predict test dataset
y_pred = clf.predict(X_test)

# Confusion matrix (without labels)
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)

# Confusion matrix with labels
target_names = ['Iris-setosa','Iris-versicolor','Iris-virginica']
cm_df = pd.DataFrame(cm, index=target_names, columns=target_names)
print("\nConfusion Matrix with Labels:\n", cm_df)

# Accuracy
print("\nAccuracy:", accuracy_score(y_test, y_pred))

# Classification report
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))

# Predict class of new record
sample = [[6.4,1.8,6.6,2.1]]
prediction = clf.predict(sample)
print("\nPredicted class for {6.4,1.8,6.6,2.1}:", prediction)

# Display tree diagram
plt.figure(figsize=(10,6))
plot_tree(clf, feature_names=X.columns, class_names=target_names, filled=True)
plt.show()

# Display tree in text form
tree_rules = export_text(clf, feature_names=list(X.columns))
print("\nDecision Tree Rules:\n")
print(tree_rules)
```

10. Apply Naïve Bayes algorithm (Gaussian Naive Bayes) to develop classifier model using 70/30 data split 
on IRIS dataset

* Develop classifier model by splitting the dataset to 70-30 of training set and testing set respectively 
* Develop classifier model using ‘GaussianNB’ 
* Predict the class label of the testing dataset 
* Calculate the accuracy of the performance on the testing dataset 
* Predict the class label of the first record of the testing dataset 
* Calculate the accuracy of the performance on the first record of the testing dataset 

```python
# Gaussian Naive Bayes on Iris dataset

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# Load dataset (IMPORTANT: assign column names)
file_path = r"C:\Users\yifru\OneDrive\Desktop\machine learning\worksheet four\iris\iris.data"

columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'Class']
data = pd.read_csv(file_path, header=None, names=columns)

# Split features and target
X = data.drop('Class', axis=1)
y = data['Class']

# 70/30 split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Model
gnb = GaussianNB()
gnb.fit(X_train, y_train)

# Predictions on test set
y_pred = gnb.predict(X_test)

print("\nPredicted classes:\n", y_pred)

# Probabilities
print("\nProbabilities:\n", gnb.predict_proba(X_test).round(3))

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"\nAccuracy: {accuracy:.2f}")

# First test record prediction
X_test1 = X_test[:1]
first_pred = gnb.predict(X_test1)

print("\nFirst record prediction:", first_pred)

print("\nFirst record probabilities:\n", gnb.predict_proba(X_test1).round(6))

# Accuracy for first record (manual check)
acc_first = accuracy_score(y_test[:1], first_pred)
print(f"\nAccuracy on first record: {acc_first:.2f}")
```

13. Apply feature selection using Information Gain ratio 
* Calculate the information gain of each feature of the iris dataset 
* Select the best 2 features of the iris dataset 

```python
# 13. Feature selection using Information Gain
import pandas as pd
import numpy as np
from sklearn.feature_selection import mutual_info_classif, SelectKBest
from sklearn.datasets import load_iris

# Load a sample dataset (Iris dataset)
file_path = r"C:\Users\yifru\OneDrive\Desktop\machine learning\worksheet four\iris\iris.data"

columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'Class']
data = pd.read_csv(file_path, header=None, names=columns)
X=data.drop('Class', axis='columns')
y=data['Class']

# Calculate mutual information (Information Gain) for each feature
mi_scores = mutual_info_classif(X, y, random_state=42)
# Create a Series to easily inspect the scores
mi_series = pd.Series(mi_scores, index=X.columns)

print("Information Gain (Mutual Information) scores for each feature:")
print(mi_series.sort_values(ascending=False))

# Select the top K features based on information gain
k_best_features = SelectKBest(mutual_info_classif, k=2) # Select top 2 features
X_selected = k_best_features.fit_transform(X, y)
print ("\n \n dataset with selected \n ", X_selected )
print("\nShape of original data:", X.shape)
print("\n Shape of data after selecting top 2 features:", X_selected.shape)

# Get the names of the selected features
selected_feature_indices = k_best_features.get_support(indices=True)
selected_feature_names = X.columns[selected_feature_indices]
print("\nSelected features:", selected_feature_names.tolist())

```
